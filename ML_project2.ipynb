# **Loading Dataset1**

```
# Importing the libraries
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from matplotlib import colors
%matplotlib inline

# Importing the dataset
col_names=['x1', 'x2', 't']
dataset1 = pd.read_csv('dataset1.csv', names=col_names, header=None)
X1= dataset1[['x1','x2']]
Y1= dataset1[['t']]

```

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X1_train, X1_test,Y1_train, Y1_test= train_test_split(X1,Y1, test_size=0.25, random_state=0)

```

# **Loading Dataset 2**

```
# Importing the dataset
col_names=['x1', 'x2', 't']
dataset2 = pd.read_csv('dataset2.csv', names=col_names, header=None)
X2= dataset2[['x1','x2']]
Y2= dataset2[['t']]
```

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X2_train, X2_test,Y2_train, Y2_test= train_test_split(X2,Y2, test_size=0.25, random_state=0)
```

# **Logistic Regression classifier on Dataset1**

```
# Fitting logistic regression to the Training set
from sklearn.linear_model import LogisticRegression
logReg1= LogisticRegression()
logReg1.fit(X1_train, Y1_train )

# Predicting the Test set results
Y1_pred_logReg= logReg1.predict(X1_test)

# Measuring Accuracy
accuracy1_logReg= logReg1.score(X1_test, Y1_test)
print("Accuracy = ", accuracy1_logReg)

# Making the Confusion Matrix
from sklearn import metrics
cnf_matrix1_logReg = metrics.confusion_matrix(Y1_test, Y1_pred_logReg)
print("Confusion Matrix")
print(cnf_matrix1_logReg)
```

Ans 1) The model predicts with 100% accuracy on test set

# **Logistic Regression classifier on Dataset 2**

```
# Fitting logistic regression to the Training set
from sklearn.linear_model import LogisticRegression
logReg2= LogisticRegression()
logReg2.fit(X2_train, Y2_train )

# Predicting the Test set results
Y2_pred_logReg= logReg2.predict(X2_test)


# Measuring Accuracy
accuracy2_logReg= logReg2.score(X2_test, Y2_test)
print("Accuracy = ", accuracy2_logReg)

# Making the Confusion Matrix
from sklearn import metrics
cnf_matrix2_logReg = metrics.confusion_matrix(Y2_test, Y2_pred_logReg)
print("Consusion matrix:  ")
print(cnf_matrix2_logReg)
```

Ans 2)  The model predicts with 60% accuracy on test set. It is less accurate as Logistic regression is linear classifier and our dataset is not linear.

# **Gaussian Naive Bayes classifier on Dataset 1**

```
from sklearn.naive_bayes import GaussianNB
gnb1 = GaussianNB()
gnb1.fit(X1_train, Y1_train)
y1_pred_gnb = gnb1.predict(X1_test)

# Measuring Accuracy
accuracy1_gnb= gnb1.score(X1_test, Y1_test)
print("Accuracy = ", accuracy1_gnb)

# Making the Confusion Matrix
from sklearn import metrics
cnf_matrix1_gnb = metrics.confusion_matrix(Y1_test, y1_pred_gnb)
print("Consusion matrix:  ")
print(cnf_matrix1_gnb)
```

# **Gaussian Naive Bayes classifier on Dataset 2**

```
from sklearn.naive_bayes import GaussianNB
gnb2 = GaussianNB()
gnb2.fit(X2_train, Y2_train)
Y2_pred_gnb = gnb2.predict(X2_test)

# Measuring Accuracy
accuracy2_gnb= gnb2.score(X2_test, Y2_test)
print("Accuracy = ", accuracy2_gnb)

# Making the Confusion Matrix
from sklearn import metrics
cnf_matrix2_gnb = metrics.confusion_matrix(Y2_test, Y2_pred_gnb)
print("Consusion matrix:  ")
print(cnf_matrix2_gnb)
```

Ans 4) Gaussian Naive bayes classifer performs better Logistic Regression classifier on dataset 2 because it is not linear classifier 

# **K Nearest Neighbor Classifier on Dataset 1**

```
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier(n_neighbors=2)
knn1.fit(X1_train, Y1_train)

Y1_pred_knn= knn1.predict(X1_test)

# Measuring Accuracy
accuracy1_knn= knn1.score(X1_test, Y1_test)
print("Accuracy = ", accuracy1_knn)

# Making the Confusion Matrix
from sklearn import metrics
cnf_matrix1_knn = metrics.confusion_matrix(Y1_test, Y1_pred_knn)
print("Consusion matrix:  ")
print(cnf_matrix1_knn)
```

# **K Nearest Neighbor Classfier on Dataset 2**

```
knn2 = KNeighborsClassifier(n_neighbors=2)
knn2 = knn2.fit(X2_train, Y2_train)
Y2_pred_knn = knn2.predict(X2_test)

# Measuring Accuracy
accuracy2_knn= knn2.score(X2_test, Y2_test)
print("Accuracy = ", accuracy2_knn)

# Making the Confusion Matrix
cnf_matrix2_knn = metrics.confusion_matrix(Y2_test, Y2_pred_knn)
print("Consusion matrix:  ")
print(cnf_matrix2_knn)
```

Ans 5) K nearest neighbour classifer performs better Logistic Regression classifier because it is not linear classifier
